## **Question 1**

## **i)**

Yes this is a valid transformation. A valid transformation is Unitary as shown. Secondly, the transformation matrix is a square matrix with valid entries.

```python
transformation = np.array([[1/np.sqrt(2), 1/np.sqrt(2), 0],
                           [-1/np.sqrt(2), 1/np.sqrt(2), 0],
                           [0, 0, 1]])

transformation_transpose = transformation.transpose()

print(f'Transpose of transformation matrix :\n{transformation_transpose}')

product = np.matmul(transformation, transformation_transpose)

print('For tranformation to be valid, it must be unitary, \n the result of the product of the matrix and its adjoint is the identity')
print('Note floating point errors that sould equal zero')
print(product)
```

The colums of the matrix repersent the new basis vectors in terms of the old basis vectors. A wavefunction ψ's new wave function can be found by multiplying the old wavefunction with the transformation matrix Ψ'(x)=TΨ(x)
