## Question 2
What is the phase difference between the states acquired by the system in t = 10−10 s?

## Awnser
P(|0⟩, t = 0.4*10**(-4)) = 0.541, as the system evolves so do the probabilities.

```python
import math
import numpy as np
from scipy.linalg import expm
from numpy.linalg import matrix_power
import scipy.constants as const
import matplotlib.pyplot as plt
%matplotlib inline

psi_0 = np.array([[1],[0]])

H = np.array([[1/np.sqrt(2), 1/np.sqrt(2)],
              [1/np.sqrt(2), -1/np.sqrt(2)]])

def Evolution(Hmatrix,t):
  return( expm((-1j*Hmatrix*t)/(const.hbar)) )

time = 0.4*10**(-4)
new_psi = np.matmul(Evolution(H,time),psi_0)

print(new_psi)
print(np.abs(new_psi[:,0])**2)

t = np.linspace(0,6,50)
psi = []
for tt in t:
    psi.append(np.matmul(Evolution(H,tt),psi_0))
psi = np.array(psi)

plt.plot(t, np.abs(psi[:,0])**2, label = "$P_0$")
plt.plot(t, np.abs(psi[:,1])**2, label = "$P_1$")
plt.legend()
plt.xlabel("t")
plt.ylabel("Probability")
plt.show()
```
